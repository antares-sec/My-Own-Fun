```python
import pandas as pd
import glob
```

# Combine the dataset

```python
# Creating path for folder
path = 'row_files' # Change this with the folder name (not the filename)
csv_files = glob.glob(path + "/*.csv")
```

```python
df_list = [pd.read_csv(file) for file in csv_files]
df_list
```

```python
# Make it one variable for all
dataset = pd.concat(df_list, ignore_index=True)
dataset.head()
```

```python
dataset.info()
```

# Saved into 1 csv file

```python
filename = 'lux.csv' # Change this to your filename
dataset.to_csv(filename)
```
