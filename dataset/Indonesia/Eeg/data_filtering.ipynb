```python
import pandas as pd
```

```python
# Load dataset
unprocs_data = pd.read_excel("sabila.xlsx")
unprocs_data.head()
```

# Processing Columns 

```python
# Removing NaN or Null values inside of dataset
def remove_nan(dataset):
    threshold = 0.5
    dataset_cleaned = dataset.loc[:, dataset.isnull().mean() < threshold]
    return dataset_cleaned
```

```python
# Removing nan columns
dataset_cleaned = remove_nan(unprocs_data)
```

```python
# Showing the result after removing columns
dataset_cleaned.info()
```

```python
# Removing row that has nan as a value
dataset_cleaned = dataset_cleaned.dropna(axis=0)
dataset_cleaned.info()
```

```python
# Change the columns head
cols = ['Date', 'Duration', 'Tag', 'Attention', 'Relaxation', 'Delta', 'Theta', 'Low-Alpha', 'High-Alpha', 'Low-Beta', 'High-Beta', 'Low-Gamma', 'Mid-Gamma', 'SyncRate', 'Time-set']
```

```python
# Change the cols name
dataset_cleaned.columns = cols
dataset_cleaned.info()
```

```python
dataset_cleaned.head()
```

# Splitting the data

```python
import os

# Define columns to explode
columns_to_explode = ['Attention', 'Relaxation', 'Delta', 'Theta', 'Low-Alpha', 'High-Alpha', 'Low-Beta', 'High-Beta', 'Low-Gamma', 'Mid-Gamma', 'SyncRate', 'Time-set']


# Prepare output directory
output_dir = 'row_files'
os.makedirs(output_dir, exist_ok=True)

# Function to explode specified columns and create a new DataFrame for each row
for index, row in dataset_cleaned.iterrows():
    # Create a new DataFrame for the current row's non-exploded data
    row_data = {'Date': [row['Date']], 'Duration': [row['Duration']], 'Tag': [row['Tag']]}
    row_df = pd.DataFrame(row_data)
    
    # Explode each specified column
    exploded_dfs = []
    max_length = 0
    for column in columns_to_explode:
        values = row[column].split(',')
        max_length = max(max_length, len(values))
        exploded_df = pd.DataFrame(values, columns=[column])
        exploded_dfs.append(exploded_df)
    
    # Extend the non-exploded data to match the length of the longest exploded column
    combined_df = pd.concat([row_df] * max_length).reset_index(drop=True)
    
    # Add exploded columns to the combined DataFrame
    for exploded_df in exploded_dfs:
        combined_df = pd.concat([combined_df.reset_index(drop=True), exploded_df.reset_index(drop=True)], axis=1)
    
    # Save to a CSV file for each row
    filename = os.path.join(output_dir, f'row_{index}.csv')
    combined_df.to_csv(filename, index=False)

print("All rows have been processed and saved as separate CSV files in the 'row_files' directory.")
```
